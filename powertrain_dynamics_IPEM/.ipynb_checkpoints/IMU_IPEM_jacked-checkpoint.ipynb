```python
# import libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%run ./scripts/util_func.py
```

```python
# import data to dataframe

data = pd.read_csv("/home/dominic/Documents/publications/sensors2021/offline_model_training/powertrain_dynamics_IPEM/data/data.csv") 

data = data.iloc[1:]

print(data)
```

```python
# convert dataframe to numpy array

array = data.to_numpy()
array =np.delete(array,[0],axis=0)

time_arr = array[:,1]
time_arr = (time_arr - time_arr[0]) * 10**(-9)
meas_index_arr = array[:,2]

left_vel_cmd_arr = array[:,3]
left_vel_meas_arr = array[:,4]
right_vel_cmd_arr = array[:,5]
right_vel_meas_arr = array[:,6]
```

```python
# plot commanded and measured wheel velocities
slice_start = 10800
slice_end = 11000

plt.figure(1, figsize=(20,8))
plt.subplot(211)
plt.plot(time_arr[slice_start:slice_end], left_vel_cmd_arr[slice_start:slice_end], c='tab:blue', label='cmd_left')
plt.plot(time_arr[slice_start:slice_end], left_vel_meas_arr[slice_start:slice_end], c='tab:orange', linestyle='dashed', label='meas_left')
plt.legend()
plt.subplot(212)
plt.plot(time_arr[slice_start:slice_end], right_vel_cmd_arr[slice_start:slice_end], c='tab:blue', label='cmd_left')
plt.plot(time_arr[slice_start:slice_end], right_vel_meas_arr[slice_start:slice_end], c='tab:orange', linestyle='dashed', label='meas_left')
plt.legend()

plt.show()
```

```python
# Run model on all data

# init tau_c and tau_d values

tau_c = 0.2
tau_d = 0.25

left_model_vel = np.zeros((meas_index_arr.shape[0]))
right_model_vel = np.zeros((meas_index_arr.shape[0]))

for i in range(10, meas_index_arr.shape[0]-1):
    
    dt = (time_arr[i] - time_arr[i-1])
    left_model_vel[i+1] = powerTrain_vel(tau_c, tau_d, left_vel_cmd_arr, left_model_vel, i, dt)
    right_model_vel[i+1] = powerTrain_vel(tau_c, tau_d, right_vel_cmd_arr, right_model_vel, i, dt)
    
```

```python
# plot commanded and measured wheel velocities

plt.figure(2, figsize=(20,8))
plt.subplot(211)
plt.plot(time_arr[slice_start:slice_end], left_vel_cmd_arr[slice_start:slice_end], c='tab:blue', label='cmd_left')
plt.plot(time_arr[slice_start:slice_end], left_vel_meas_arr[slice_start:slice_end], c='tab:orange', linestyle='dashed', label='meas_left')
plt.plot(time_arr[slice_start:slice_end], left_model_vel[slice_start:slice_end], c='tab:green', linestyle='dashdot', label='mod_left')
plt.legend()
plt.subplot(212)
plt.plot(time_arr[slice_start:slice_end], right_vel_cmd_arr[slice_start:slice_end], c='tab:blue', label='cmd_left')
plt.plot(time_arr[slice_start:slice_end], right_vel_meas_arr[slice_start:slice_end], c='tab:orange', linestyle='dashed', label='meas_left')
plt.plot(time_arr[slice_start:slice_end], right_model_vel[slice_start:slice_end], c='tab:green', linestyle='dashdot', label='mod_left')
plt.legend()

plt.show()
```

```python
# Run EKF for model parameters

tau_c = 0.5
tau_d = 0.3

left_model_vel = np.zeros((meas_index_arr.shape[0]))
#right_model_vel = np.zeros((meas_index_arr.shape[0]))
pred_start = 0

d_tau_c = 0.15
tau_c_add = tau_c + d_tau_c
tau_c_sub = tau_c - d_tau_c

d_tau_d = 0.15
tau_d_add = tau_d + d_tau_d
tau_d_sub = tau_d - d_tau_d

vel_tau_c_add = 0
vel_tau_c_sub = 0

vel_tau_d_add = 0
vel_tau_d_sub = 0

P = np.array([100.0, 100.0])
trans_mat = np.array([0.0,0.0])
Cv = 0.1

tau_arr = np.zeros((int(meas_index_arr[-1]+1), 2))
tau_arr[0,0] = tau_c
tau_arr[0,1] = tau_d

I_2 = np.eye(2,2)

for i in range(2, meas_index_arr.shape[0]-1): 
    dt = time_arr[i] - time_arr[i-1]
    
    left_model_vel[i+1] = powerTrain_vel(tau_c, tau_d, left_vel_cmd_arr, left_model_vel, i, dt)
    
    vel_tau_c_add += powerTrain_vel(tau_c_add, tau_d, left_vel_cmd_arr, left_model_vel, i, dt)
    vel_tau_c_sub += powerTrain_vel(tau_c_sub, tau_d, left_vel_cmd_arr, left_model_vel, i, dt)
    
    vel_tau_d_add += powerTrain_vel(tau_c, tau_d_add, left_vel_cmd_arr, left_model_vel, i, dt)
    vel_tau_d_sub += powerTrain_vel(tau_c, tau_d_sub, left_vel_cmd_arr, left_model_vel, i, dt)
    
    if meas_index_arr[i+1] != meas_index_arr[i]:
        #update EKF
        trans_mat[0] = central_diff(vel_tau_c_add, vel_tau_c_sub, d_tau_c)
        trans_mat[1] = central_diff(vel_tau_d_add, vel_tau_d_sub, d_tau_d)
        
#         print(trans_mat)
        
        P = np.dot(np.dot(trans_mat.T, P), trans_mat) + Cv # command noise missing
#         print(P)
        
        innov = left_vel_meas_arr[i+1] - left_model_vel[i+1]
        K_gain = np.dot(np.dot(P, trans_mat), np.dot(trans_mat, np.dot(P, trans_mat)))
        
        tau_arr[int(meas_index_arr[i+1]), :] = tau_arr[int(meas_index_arr[i])] + np.dot(K_gain, innov)
        tau_c = tau_arr[int(meas_index_arr[i]), 0]
        tau_d = tau_arr[int(meas_index_arr[i]), 1]
        
        if np.abs(tau_c) > 1.0 or np.isnan(tau_c):
            tau_c = tau_arr[int(meas_index_arr[i])-1, 0]
            tau_arr[int(meas_index_arr[i]), 0] = tau_c
        if np.abs(tau_d) > 1.0 or np.isnan(tau_d):
            tau_d = tau_arr[int(meas_index_arr[i])-1, 1]
            tau_arr[int(meas_index_arr[i]), 1] = tau_d
        
        P = np.dot(I_2 - np.dot(K_gain, trans_mat), P)
        
        pred_start = i
        vel_tau_c_add = left_vel_meas_arr[i+1]
        vel_tau_c_sub = left_vel_meas_arr[i+1]
        vel_tau_d_add = left_vel_meas_arr[i+1]
        vel_tau_d_sub = left_vel_meas_arr[i+1]
        
```
